### 3) Mecab 실행하기

```python
# MeCab 불러오기
import MeCab
m = MeCab.Tagger()
```

```python
# 기본 출력 결과
print(m.parse(text3))
```

```python
# ""(또는 '')안에 직접 문장 입력하여 출력 형태 바꿔 보기+전처리 필요
# import MeCab
# m = MeCab.Tagger()
data=m.parse("우리 동네 담배 가게에는 아가씨가 예쁘다네.")
data=data.replace("EOS","")     # 모르는 오류 삭제
data=data.rstrip()              # 우측 공백 삭제
# print(data)
# print("- - "*20)
# # data.split("\n")            # 줄바꿈 단위로 리스트 만들기
word=[]
for i in data.split("\n") :   # 줄바꿈으로 생성된 리스트 하나씩 읽기
  data=i.split(',')[0].split('	')
  word.append(data)
# 출력 형식에 따라 결과를 다양화할 수 있다.(텍스트마이닝, 전처리 과정에서 필요)
print(word)
print(word[0][0],word[0][1])
print(word[1][0],word[1][1])
```

```python
# 데이터프레임 형식으로 출력하기
import pandas as pd
column_name = ['형태소','품사']
df=pd.DataFrame(word, columns=column_name)
df
```

### 4) 텍스트 분석 연습하기 (문재인 대통령 취임사)

```python
# 파일 읽기
with open("/content/문재인대통령취임사.txt", "r", encoding = "utf-8-sig") as f:
  text=f.read()
  # print(text)
```

```python
# 전처리(1) 후 데이터프레임으로 출력
data=m.parse(text)
data=data.replace("EOS","")   # 모르는 오류 삭제
data=data.rstrip()            # 우측 공백 삭제
# print(data)
word=[]
for i in data.split("\n") :   # 줄바꿈으로 생성된 리스트 하나씩 읽기
  data=i.split(',')[0].split('	')
  word.append(data)
print(word)
column_name = ['형태소','품사']
df=pd.DataFrame(word, columns=column_name)
df
```

```python
# 빈도 분석을 위한 전처리
text_filter=str(word)
text_filter=text_filter.replace("[['","")
text_filter=text_filter.replace("']]","")
text_filter=text_filter.replace("'], ['"," ")
text_filter=text_filter.replace("', '","/")
print(text_filter)

f=open("/content/취임사_pos.txt",'w')  # 형태소 분석 결과를 저장한 파일을 생성한다.
f.write(text_filter)
f.close()
```

```python
with open("/content/취임사_pos.txt", "r", encoding = "utf-8") as inp:
    product = dict()
    for p in inp.read().split():
        # for p in line.split("+"):
            if (p in product):
                product[p] += 1
            else:
                product[p] = 1
    with open("/content/취임사_frq.txt", "w", encoding = "utf-8") as outp:
        for p, v in sorted(product.items(), key = lambda x: x[1], reverse = True):
            print("{}({})".format(p, v))
            outp.write("{}({})\n".format(p, v))
```
