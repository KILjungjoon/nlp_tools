```python
import pandas as pd
from jamo import h2j, j2hcj

# 자모음 음소가 포함된 임의의 DataFrame을 만들자
df = pd.DataFrame({'headword': ['한글', 'ㄱ', '히', 'ㅢ', '기쁨', 'ㅞ', '가', '매일', 'ㄺ', '커피', '시작']})
df
```

```python
# h2j, j2hcj 메소드를 사용하여 음소 분리 후 단어의 길이가 2 이상인지 확인하는 함수 정의
def has_multiple_phonemes(word):
    return len(j2hcj(h2j(word))) >= 2

# 함수를 적용하여 자모음을 걸러내고 음절만 저장
df = df[df['headword'].apply(has_multiple_phonemes)]

# 인덱스 리셋
df.reset_index(drop=True, inplace=True)

df
```
